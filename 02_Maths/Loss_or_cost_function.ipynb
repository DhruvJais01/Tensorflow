# Implementation of loss function


```python
import numpy as np

y_predicted = np.array([1, 1, 0, 0, 1])
y_true = np.array([0.30, 0.7, 1, 0, 0.5])
```

```python
def mae(y_true, y_predicted):
    total_loss = 0
    for yt, yp in zip(y_true, y_predicted):
        total_loss += abs(yt - yp)
    print("Total loss: ", total_loss)
    mae = total_loss / len(y_true)
    print("MAE: ", mae)
    return mae
```

```python
mae(y_true, y_predicted)
```

```python
def log_loss(y_true, y_predicted):
    eplion = 1e-15
    y_predicted_new = [max(i, eplion) for i in y_predicted]
    y_predicted_new = [min(i, 1-eplion) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true * np.log(y_predicted_new) + (1 - y_true) * np.log(1 - y_predicted_new))
```

```python
log_loss(y_true, y_predicted)
```

**Mean Squared Error (MSE)**


```python
import numpy as np

y_predicted = np.array([1, 1, 0, 0, 1])
y_true = np.array([0.30, 0.7, 1, 0, 0.5])
```

```python
np.mean(np.square(y_predicted - y_true))  # mean squared error
```

```python
def mse(y_true, y_predicted):
    total_loss = 0
    for yt, yp in zip(y_true, y_predicted):
        total_loss += (yt - yp) ** 2
    print("Total loss: ", total_loss)
    mse = total_loss / len(y_true)
    print("MSE: ", mse)
    return mse
```

```python
mse(y_true, y_predicted)
```
