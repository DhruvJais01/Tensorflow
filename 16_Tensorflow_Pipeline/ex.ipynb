## Retrieve review file paths in a tensorflow dataset

```python
import pandas as pd

df_neg = tf.data.Dataset.list_files("train_neg.csv")
# Read the CSV files using pandas
df_neg = pd.read_csv("train_neg.csv", header=None, skiprows=1).sample(20, random_state=1)
df_pos = pd.read_csv("train_pos.csv", header=None, skiprows=1).sample(20, random_state=1)

# Extract text from the dataframes
neg_reviews = df_neg[0].tolist()
pos_reviews = df_pos[0].tolist()

# Combine the reviews
reviews_ds = neg_reviews + pos_reviews

reviews_ds_tf = tf.data.Dataset.from_tensor_slices(reviews_ds)
def label_review(review):
    if review in neg_reviews:
        return 'negative', review
    elif review in pos_reviews:
        return 'positive', review
    else:
        return 'unknown', review

# Apply the function to the dataset
labeled_reviews = reviews_ds_tf.map(lambda review: tf.py_function(func=label_review, inp=[review], Tout=[tf.string, tf.string]))

for label, review in labeled_reviews.take(10):
    print(f"{label}, Review: {review.numpy()}")
```

```python
reviews_ds_2 = labeled_reviews.filter(lambda review, label: review != b"")
for review, label in reviews_ds_2.as_numpy_iterator():
    print("Review: ", review[:50].decode('utf-8'))
    print("Label: ", label.decode('utf-8'))
```

```python
final_ds = reviews_ds_tf.map(lambda review: tf.py_function(func=label_review, inp=[review], Tout=[tf.string, tf.string])).filter(lambda review, label: review != b"").shuffle(3)
```

```python
for review, label in final_ds.as_numpy_iterator():
    print("Review:",review[:50])
    print("Label:",label)
```
